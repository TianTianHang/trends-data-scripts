```python
import pandas as pd


# 读取networkuse.csv
networkuse_df = pd.read_csv('data/networkuse.csv')

# 读取LatinTimeline.csv
latin_timeline_df = pd.read_csv('data/LatinTimeline.csv')

# 读取population.csv
population_df = pd.read_csv('data/population.csv')

# 读取geo数据
world = pd.read_csv("data/countries.csv")
```

```python
index=[0,1]
time_index=range(48,68)
index.extend(range(48,68))
#04-23 使用网络人数占比
networkuse_df=networkuse_df.iloc[...,index]
population_df=population_df.iloc[...,index]
#数据缺失超过9年
drop_index=[0, 11, 38, 51, 84, 108, 110, 147, 164, 179, 188, 193, 206, 216, 225, 228,
       255, 261]
networkuse_df.drop(index=drop_index,inplace = True)
population_df.drop(index=drop_index,inplace = True)
networkuse_df.interpolate(method='linear', axis=0,inplace = True)
```

```python
#使用网络人数
result=networkuse_df.iloc[...,2:].mul(population_df.iloc[...,2:])/100
population_networkuse_df=pd.concat([networkuse_df.iloc[...,[0,1]],result],axis=1)
population_networkuse_df
```

```python
population_networkuse_df.to_csv("data/population_networkuse.csv",index=False)
```
